```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Sorting and Filtering

```python
ds_1e3 = pd.read_csv(".result/s1000/sort-filter.csv")
ds_1e4 = pd.read_csv(".result/s10000/sort-filter.csv")
ds_1e5 = pd.read_csv(".result/s100000/sort-filter.csv")
```

```python
pd.DataFrame([
    list(ds_1e3.drop(["test_no"], axis=1).mean()) + [len(ds_1e3)],
    list(ds_1e4.drop(["test_no"], axis=1).mean())+ [len(ds_1e4)],
    list(ds_1e5.drop(["test_no"], axis=1).mean())+ [len(ds_1e5)]
], 
index=["s1000", "s10000", "s100000"],
columns=["Sorting (ms)", "Filtering (ms)", "Count"]
)
```

### Minimum Coin Change (Dynamic Programming)

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
test_coin_change = pd.read_csv(".result/coin-change.csv", index_col=0)
test_coin_change["min_coin"].plot(
    ax=ax[0],
    ylabel="Minimum coins",
    title="Minimum Coins vs Value"
)
test_coin_change["exec_time"].plot(
    ax=ax[1],
    ylabel="Execution time (ms)",
    title="Exectution Time vs Value",
    color="orange",
)
plt.savefig("../../Images/min-coin-change")
```
